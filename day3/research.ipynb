```python
df3 = pd.read_json('lastdata_papers_with_code.json')
```

```python
dictionary = dict()
for i in range(len(df3)):
    if df3["task"][i]:
        for task in df3["task"][i]:
            #print(task)
            if task in dictionary:
                dictionary[task] += 1
            else:
                dictionary[task] = 1   
```

```python
dictionary
```

```python
frame = {"unknown":0}
for i in range(len(df3)):
    if df3["frame"][i]:
        for task in df3["frame"][i]:
            #print(task)
            if task in frame:
                frame[task] += 1
            else:
                frame[task] = 1
    else:
        frame["unknown"] += 1
```
