```python
import pandas as pd
df1 = pd.read_json('lastdata_papers_with_code.json')
```

```python
dictionary1 = dict()
for i in range(len(df1)):
    if df1["task"][i]:
        for task in df1["task"][i]:
            #print(task)
            if task in dictionary1:
                dictionary1[task] += 1
            else:
                dictionary1[task] = 1   
```

```python
dictionary1
```

```python
frame1 = {"unknown":0}
for i in range(len(df1)):
    if df1["frame"][i]:
        for task in df1["frame"][i]:
            #print(task)
            if task in frame1:
                frame1[task] += 1
            else:
                frame1[task] = 1
    else:
        frame1["unknown"] += 1
```

```python
frame1
```
